## Independent T-test

The Student’s t-test is a statistical test that compares the mean and standard deviation of two samples to see if there is a significant difference between them.  In an experiment, a t-test might be used to calculate whether or not differences seen between the control and each experimental group are a factor of the manipulated variable or simply the result of chance.

## Data

```python
import numpy as np
#group_id #value
data = [[0, 7],[0, 10],[0, 6],[0, 5],[0, 5],[0, 6],[0, 12],[0, 6],[0, 11],[0, 6],[1, 3],[1, 1],[1, 0],[1, 4],[1, 2],
        [1, 5],[1, 3],[1, 2],[1, 3],[1, 4]]

data_np = np.array(data)
```

## Hypothesis

Null Hypothesis:
"There is not a significant difference between the two groups; any observed differences may be due to chance and sampling error."

Alternative Hypothesis:
"There is a significant difference between the two groups; the observed differences are most likely not due to chance or sampling error."

## Test Statistics 

```python
data_g0 = data_np[data_np[:,0]==0]
data_g1 = data_np[data_np[:,0]==1]
x1 = data_g0[:,1].mean()
x2 = data_g1[:,1].mean()
n1 = len(data_g0[:,1])
n2 = len(data_g1[:,1])
s1 = data_g0[:,1].std(ddof=1)
s2 = data_g1[:,1].std(ddof=1)
diff_mean = x1-x2
avg_std = (s1**2/n1+s2**2/n2)**0.5
ttest = diff_mean/avg_std
print('T-test: ',ttest)
```

## P-value

```python
from scipy.stats import t
dof = n1+n2-2
pvalue = t.sf(ttest, dof)*2 # two-sided pvalue
print("Two-sided p-value: ",pvalue)
```

## Decision

The calculated P-value is less than the significance level of 0.05, and therefore, the null hypothesis can be rejected. The data suggests that there is a significant difference between the two groups.
